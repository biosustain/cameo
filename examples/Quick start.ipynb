Step 1: Load a model
-------------------

Loading a model is easy. From the cameo package you can import the *load_model* function.

```python
from cameo import load_model
```

```python
print(load_model.__doc__)
```

```python
model = load_model("iJO1366")
```

```python
model
```

```python
print(model.objective)
```

Step 2: Optmize the mathematical problem (solve the model)
------------------------------------------------------

The current objective can be optmized by calling the *solve()* or *optimize()* methods.
solve() will raise a ***SolveException*** when the problem is infeasible.

The sense of the problem (or direction) is described the objective itself.

```python
solution = model.solve()
```

```python
solution
```

Step 3: The metabolic model
--------------------------

The **SolverBasedModel** object allows to explore directly explore certain properties of the metabolic model:

* *reactions*: the reactions contained in the model (it supports tab-complete; with some minor limitations...)


```python
model.reactions
```

```python
model.reactions.E4PD
```

* *exchanges*: reactions that allow the exchange of metabolites between the environment and the model

```python
model.exchanges
```

* *metabolites*: the metabolic species present in the model (also supports tab-complete)

```python
model.metabolites.get_by_id("12dgr140_p")
```

```python
model.metabolites.glc_dsh_D_c
```

* *medium*: the current limites for exchange rates (other then 0 to 0)

```python
model.medium()
```

* *genes*: the genes (if present) associated with the reactions in the model

```python
model.genes
```

* *essential_reactions*: reactions that cannot be knocked-out using the current constraints to achieve current objective

```python
model.essential_reactions()
```

* *essential_genes*: genes that cannot be knocked-out using the current constraints to achieve the current objective

```python
model.essential_genes()
```
