# Visbio and cobrapy

cobrapy integrates well with the [visbio](https://github.com/zakandrewking/visbio) package. If it has been installed, viewing maps is extremely simple. The default map is the _E. coli_ core map, but other maps can be viewed by passing the correct map_name. The maps are then downloaded from a map [repository](https://github.com/zakandrewking/visbio/tree/gh-pages/maps).

```python
def show(bound=20.):
    model.reactions.get_by_id('EX_o2_e').lower_bound = -1*bound
    model.optimize()
    wt_solution = model.solution.x_dict
    display(wt_solution['Ec_biomass_iJO1366_core_53p95M'])
    display(HTML(Builder('iJO1366_central_metabolism',reaction_data=wt_solution)._get_html()))
#     display(Map(flux={'ENO':upper_obj_bound,'PGK':2}))
```

```python
i = interact(show, bound=widgets.FloatSliderWidget(min=0.0, max=20.0, step=0.1, value=1.0))
```
